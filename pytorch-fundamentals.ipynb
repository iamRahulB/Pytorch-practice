```python
data=[
    [[2,3,4],[2,3,4]]
]

#we can think of tensor is simple a
tensor=torch.tensor(data)

np_data=np.array(data)
tensor.ndim
print(np_data)
print(torch.tensor(np_data))

```

```python
print(torch.ones_like(tensor),"\n")

print(torch.ones_like(tensor,dtype=torch.float),"\n")


print(torch.rand_like(tensor,dtype=torch.float),'\n')

# it copies property of the already made tensor


# we can define shape to generate the tensor also

print(torch.zeros((2,3),dtype=torch.int))

print(torch.rand((2,3)))
```
